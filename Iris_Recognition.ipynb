```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris=load_iris()
iris
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df["species"]=iris.target
df.sample(10)
```

```python
df["species"].value_counts()
```

```python
df.describe()
```

```python
f=df.drop(["species"],axis='columns')
f
```

```python
import seaborn as sn
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(f,df.species,test_size=0.2)
```

```python
import matplotlib.pyplot as plt
```

```python
g=sn.pairplot(df,hue='species',markers='+',palette="tab10")
#sn.set(style="ticks", color_codes=True)
plt.show()
```

```python
ax = sn.swarmplot(x='sepal length (cm)', y='species', data=df,orient="h")
```

```python
s=sn.violinplot(y="species",x="sepal length (cm)",data=df,inner='quartile',orient='h')
plt.show()
s=sn.violinplot(y="species",x="sepal width (cm)",data=df,inner='quartile',orient='h')
plt.show()
s=sn.violinplot(y="species",x="petal length (cm)",data=df,inner='quartile',orient='h')
plt.show()
s=sn.violinplot(y="species",x="petal width (cm)",data=df,inner='quartile',orient='h')
plt.show()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.predict(x_test)
```

```python
lg.score(x_test,y_test)
```

```python
y_predicted=lg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(y_test,y_predicted)
```

```python
sn.heatmap(cf,annot=True)
plt.xlabel("features")
plt.ylabel("species")
plt.show
```
